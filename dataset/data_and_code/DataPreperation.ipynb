# Importing the libraries

```python
import numpy as np
import pandas as pd
import os
```

```python
# function which reads the data from text files and returns a dataframe
def makedataframe(path,value):
    review=[]
    for i in os.listdir(path):
        with open(os.path.join(path,i),'r') as f:
            temp = f.read()
            review.append(''.join(temp.split('\n')))
    f.close()
    label = [value for i in review]
    df=pd.DataFrame(np.column_stack([review,label]),
                   columns=['Review','labels'])
    return df
```

```python
positive = os.getcwd()+'\\pos' 
negative = os.getcwd()+'\\neg'
pos = makedataframe(positive,0)
neg = makedataframe(negative,1)
```

```python
df = pd.concat([pos, neg], ignore_index=True)
```
